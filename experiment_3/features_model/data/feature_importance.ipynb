```python
from sklearn.ensemble import RandomForestClassifier
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# define a function to select features
def select_features(data, labels, top_num=20):
    model = RandomForestClassifier(random_state=42)
    model.fit(data, labels)
    importances = model.feature_importances_
    indices = np.argsort(importances)[::-1]
    selected_features = [data.columns[i] for i in indices[:top_num]]
    
    # plot the feature importances
    plt.figure(figsize=(10, 8))
    plt.title("Feature importances")
    plt.bar(range(top_num), importances[indices[:top_num]], color="r", align="center")
    plt.xticks(range(top_num), selected_features, rotation=90)
    plt.xlim([-1, top_num])

    return selected_features

# read train data
df_train = pd.read_csv('train_set_850-870_features.csv')
train_data = df_train.drop(['mean', 'label'], axis=1)

# read test data
df_test = pd.read_csv('test_set_850-870_features.csv')
test_data = df_test.drop(['mean', 'label'], axis=1)
```

```python
# select features (the importance of 31st is over 0.015)
selected_features = select_features(train_data, df_train['label'], 30)
print(selected_features)
plt.show()
```

```python
# extract selected features from the original train data
df_train_selected = df_train[selected_features]
df_train_selected['label'] = df_train['label']
df_train_selected['mean'] = df_train['mean']

# save the selected features to a new csv file
df_train_selected.to_csv('train_set_850-870_top30_features.csv', index=False)
```

```python
# extract selected features from the original test data
df_test_selected = df_test[selected_features]
df_test_selected['label'] = df_test['label']
df_test_selected['mean'] = df_test['mean']

# save the selected features to a new csv file
df_test_selected.to_csv('test_set_850-870_top30_features.csv', index=False)
```
